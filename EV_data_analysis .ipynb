```python
data['County'].fillna('Unknown', inplace=True)
```

```python
data['City'].fillna('Unknown', inplace=True)
```

```python
data['Electric Utility'].fillna('Unknown', inplace=True)
```

```python
# Fill missing numeric values with mean or median
data['Postal Code'].fillna(data['Postal Code'].mode()[0], inplace=True)
```

```python
data['Legislative District'].fillna(data['Legislative District'].median(), inplace=True)
```

```python
data['2020 Census Tract'].fillna(data['2020 Census Tract'].median(), inplace=True)
```

```python
# Drop rows with critical missing values (if necessary)
data.dropna(subset=['Vehicle Location'], inplace=True)
data.isnull().sum()
```

```python
# Data Manipulation
data['State'] = data['State'].str.upper()
```

```python
data['Electric Vehicle Type'] = data['Electric Vehicle Type'].str.title()
```

```python
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# removing duplicates
data.drop_duplicates(inplace=True)
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Check for invalid Electric Range values
print(data[data['Electric Range'] < 0])  # Identify rows with negative range
```

```python
# Replace invalid Electric Range values
data['Electric Range'] = data['Electric Range'].apply(lambda x: max(x, 0))

# Validate Model Year
data = data[(data['Model Year'] >= 2000) & (data['Model Year'] <= 2025)]
```

```python
data
```

```python
# Convert Postal Code to integer (assuming it represents zip codes)
data['Postal Code'] = data['Postal Code'].astype(int)

# Convert Legislative District to integer
data['Legislative District'] = data['Legislative District'].astype(int, errors='ignore')

# Optimize categorical columns
categorical_columns = ['County', 'City', 'State', 'Make', 'Model', 'Electric Vehicle Type', 
                       'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Electric Utility']
data[categorical_columns] = data[categorical_columns].astype('category')
```

```python
# renaming
data.rename(columns={
    'VIN (1-10)': 'VIN',
    'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV Eligibility',
    'Base MSRP': 'Base Price',
    '2020 Census Tract': 'Census Tract'
}, inplace=True)
```

```python
# Check for duplicate VINs
duplicate_vins = data['VIN'].duplicated().sum()
print(f"Number of duplicate VINs: {duplicate_vins}")

# Drop rows with duplicate VINs
data.drop_duplicates(subset='VIN', inplace=True)
duplicate_vins = data['VIN'].duplicated().sum()
```

```python
# Visualize Base MSRP for outliers
sns.boxplot(data['Base Price'])
plt.show()

# Remove outliers beyond a threshold
q1 = data['Base Price'].quantile(0.25)
q3 = data['Base Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

data = data[(data['Base Price'] >= lower_bound) & (data['Base Price'] <= upper_bound)]
data
```

```python
# grouping and inspecting data by CAFV Eligibility 
var=data.groupby('CAFV Eligibility')
for x in var:
    print(x)
    
```

```python
x=pd.melt(data)
x[5:15]
```

```python
# adding new column
data['Price per Mile']=data['Base Price']/data['Electric Range']
data
```

```python
# adding new column using VIN
data['Manufacturer Code'] = data['VIN'].str[:2]
data
```

```python
data['Is Luxury'] = data['Base Price'] > 50000
data.head()
```

```python
# Agrregate data of variuous metrices of data 
aggregated_data = data.groupby(['Make', 'Model Year']).agg(
    Avg_MSRP=('Base Price', 'mean'),
    Avg_Electric_Range=('Electric Range', 'mean'),
    Total_Vehicles=('VIN', 'count')
).reset_index()

print(aggregated_data)
```

```python
# adding new column of high population by census tract 
data['High Population Area'] = data['Census Tract'] > data['Census Tract'].median()
data
```

```python
# Creating a column for the year category (e.g., old vs. new vehicles)
data['Age Category'] = pd.cut(
    2025 - data['Model Year'], bins=[0, 5, 10, 20], labels=['0-5 years', '6-10 years', '11+ years']
)

# Group by Electric Utility and calculate vehicle counts
utility_data = data.groupby('Electric Utility').agg(
    Total_Vehicles=('VIN', 'count')
).sort_values(by='Total_Vehicles', ascending=False).reset_index()

print(utility_data)
```

```python
data
```

```python
# Counting vehicles per year to observe growth
yearly_trends = data.groupby('Model Year').agg(
    Total_Vehicles=('VIN', 'count')
).reset_index()

# Calculating year-over-year growth rate
yearly_trends['YoY Growth (%)'] = yearly_trends['Total_Vehicles'].pct_change() * 100

print(yearly_trends)
```

```python
# Defineing price range bins and labels
price_bins = [0, 30000, 60000, 100000, float('inf')]
price_labels = ['Budget', 'Mid-Range', 'Premium', 'Luxury']

# Adding a column for price segments
data['Price Segment'] = pd.cut(data['Base Price'], bins=price_bins, labels=price_labels, right=False)
data
```

```python
# Counting the most popular models for each state
popular_models_by_state = data.groupby(['State', 'Model']).agg(
    Vehicle_Count=('VIN', 'count')
).sort_values(by=['State', 'Vehicle_Count'], ascending=[True, False]).reset_index()

# Filtering top models for each state
top_models_by_state = popular_models_by_state.groupby('State').head(3)
print(top_models_by_state)
```

```python
# Counting vehicles by postal code
density_by_postal = data.groupby('Postal Code').agg(
    Total_Vehicles=('VIN', 'count')
).reset_index()

# Adding density category based on total vehicles
density_bins = [0, 50, 200, 500, float('inf')]
density_labels = ['Low', 'Medium', 'High', 'Very High']
density_by_postal['Density Category'] = pd.cut(
    density_by_postal['Total_Vehicles'], bins=density_bins, labels=density_labels, right=False
)

print(density_by_postal)
```

```python
# Group by vehicle type and calculate average electric range
range_by_type = data.groupby('Electric Vehicle Type').agg(
    Avg_Electric_Range=('Electric Range', 'mean'),
    Total_Vehicles=('VIN', 'count')
).reset_index()

print(range_by_type)
```

```python
# Group by electric utility and calculate total vehicles
utility_coverage = data.groupby('Electric Utility').agg(
    Total_Vehicles=('VIN', 'count')
).sort_values(by='Total_Vehicles', ascending=False).reset_index()

print(utility_coverage)
```

```python
# Calculate vehicle age
data['Vehicle Age'] = 2025 - data['Model Year']

# Categorize vehicles into age groups
age_bins = [0, 5, 10, 15, 20, float('inf')]
age_labels = ['0-5 years', '6-10 years', '11-15 years', '16-20 years', '20+ years']
data['Age Group'] = pd.cut(data['Vehicle Age'], bins=age_bins, labels=age_labels, right=False)

# Group by age group and count vehicles
vehicles_by_age_group = data.groupby('Age Group').agg(
    Total_Vehicles=('VIN', 'count')
).reset_index()

print(vehicles_by_age_group)
```

```python
# Group by state and count vehicles
top_states = data.groupby('State').agg(
    Total_Vehicles=('VIN', 'count')
).sort_values(by='Total_Vehicles', ascending=False).reset_index()

# Filter top 5 states
top_5_states = top_states.head(5)
print(top_5_states)
```

CONCLUSION :
Insights From Data :
1. Popular EV Manufacturers
Insight: Identify the top manufacturers of electric vehicles (e.g., Tesla, Hyundai, Jeep).
Tesla likely dominates with high-range electric vehicles like the Model 3, whereas Jeep specializes in Plug-in Hybrid Electric Vehicles (PHEVs).

Business Action: Focus on partnerships with top-performing manufacturers or promote their EV models in specific regions.

2. Electric Range Trends by Model Year
Insight: EVs manufactured in recent years (2022–2023) show improved electric ranges compared to older models.

Business Action: Highlight newer models with better electric range in marketing campaigns to attract range-conscious customers.

3. Clean Alternative Fuel Vehicle (CAFV) Eligibility
Insight: A significant number of Plug-in Hybrid Electric Vehicles (PHEVs) are ineligible for CAFV due to low battery range.

Business Action: Educate manufacturers and customers about CAFV eligibility criteria to promote cleaner alternatives and improve vehicle designs for compliance.

4. Geographic Concentration of EVs
Insight: King County (Seattle area) has the highest concentration of EVs, followed by Yakima and Thurston Counties.

Business Action:
Focus charging station infrastructure investments in areas with high EV adoption.
Create region-specific marketing campaigns to target EV users in these counties.

5. Popular Electric Vehicle Types
Insight: Battery Electric Vehicles (BEVs) dominate over Plug-in Hybrid Electric Vehicles (PHEVs) in terms of numbers and range.

Business Action: Promote BEVs as the preferred choice for environmentally-conscious customers. Offer incentives or rebates for PHEV users to transition to BEVs.

6. Price Analysis (Base MSRP vs. Electric Range)
Insight: EVs with higher electric ranges typically come at a premium price (higher MSRP). However, certain brands (e.g., Hyundai Kona) offer good electric range at competitive prices.

Business Action:
Develop cost-effective EV financing programs for customers.
Promote models with the best range-to-cost ratio for value-driven customers.

7. Legislative District Impact
Insight: Some districts have a high density of EVs, indicating better adoption due to supportive policies or charging infrastructure.

Business Action: Collaborate with policymakers in underperforming districts to promote EV adoption via subsidies, tax incentives, or improved infrastructure.

8. Electric Utility Coverage
Insight: Some utilities, such as "PUGET SOUND ENERGY INC" and "CITY OF SEATTLE," cater to multiple regions with EV support, whereas others are localized.

Business Action: Partner with dominant electric utility providers to improve EV charging availability and reliability in underserved regions.

9. Vehicle Location Insights
Insight: Geographic clustering of vehicle locations highlights urban areas as primary hubs for EV adoption.

Business Action:
Prioritize EV dealerships and charging station placements in urban clusters.
Provide mobile charging solutions in rural areas to encourage EV adoption.

10. EV Adoption by Postal Code
Insight: Certain postal codes show higher EV registrations, correlating with higher income levels and eco-conscious populations.

Business Action:
Use targeted advertising in these postal codes.
Offer referral bonuses to existing EV owners in these regions to attract more customers.
